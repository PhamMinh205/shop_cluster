```python
import pandas as pd
from src.cluster_library import RuleBasedCustomerClusterer

transactions = pd.read_csv("data/transactions.csv")
rules = pd.read_csv("data/rules.csv")
```

```python
clusterer = RuleBasedCustomerClusterer(transactions, rules)
```

```python
clusterer.build_customer_item_matrix()

X_rule_binary = clusterer.build_rule_feature_matrix(
    top_k=20,
    weighted=False
)
```

```python
X_rule_weighted = clusterer.build_rule_feature_matrix(
    top_k=20,
    weighted=True,
    weight_col="lift"
)
```

```python
scores = clusterer.choose_k_by_silhouette(range(2, 11))
scores
```

```python
cluster_labels = clusterer.fit_kmeans(k=4)
```

```python
import matplotlib.pyplot as plt

X_2d = clusterer.project_2d()

plt.scatter(X_2d[:,0], X_2d[:,1], c=cluster_labels)
plt.title("Customer Clusters (PCA)")
plt.show()
```

```python
cluster_summary = clusterer.profile_clusters()
cluster_summary
```
